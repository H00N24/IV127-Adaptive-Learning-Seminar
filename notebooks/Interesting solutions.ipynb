```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
```

```python
code_frame = pd.read_csv("../data/code_frame.csv", header=0)
solutions_frame = pd.read_csv("../data/solutions_frame.csv", header=0)
```

```python
interesting = [
    (73, 74, "Vnořené čtverce"),
    (72, 73, "Velká šachovnice"),
    (70, 71, "Sudoku řádek"),
    (68, 69, "Rozklad na prvočísla"),
    (65, 66, "Kámen, nůžky, papír"),
    (56, 57, "Sudá čísla"),
    (55, 56, "Čísla pozpátku"),
    (52, 53, "Prostřední z trojice"),
    (42, 43, "Lichá a sudá"),
    (38, 39, "Ohrožení dámou"),
    (29, 30, "Nejčastější prvek"),
    (28, 29, "Unikátní prvky"),
    (27, 28, "Zipování řetězců"),
    (22, 23, "Největší společný dělitel"),
]

interesting_id = {i[0] for i in interesting}
```

```python
for solution_row in solutions_frame.itertuples():
    if solution_row[0] not in interesting_id:
        continue
    selected_frame = code_frame[code_frame["item"] == solution_row[1]]
    if selected_frame.shape[0] == 0:
        continue

    id_cols = [
        "id",
        "item",
        "correct",
        "dec_answer",
        "ast_clean",
    ]

    data_cols = [
        x
        for x in selected_frame.columns
        if x not in id_cols and np.any(selected_frame[x] != 0)
    ]

    isoforest = IsolationForest()
    predicted = isoforest.fit_predict(selected_frame[data_cols])

    scores = isoforest.score_samples(selected_frame[data_cols])

    print(solution_row[2])
    print(solution_row[3])
    print("-" * 20)

    for _, id, correct, solution in selected_frame.iloc[np.argsort(scores)[:5]][
        ["id", "correct", "dec_answer"]
    ].itertuples():
        print(id, correct, solution)
        print()
```
